## 算法思路

### **算法思路概述**

1. **问题理解**：
    - **目标**：从两个起点路口 `A` 和 `B` 出发，选择同一个方向（X 或 Y）行走，找到使两者所在路口类型不同的最少步数（即“路口独特度”）。
    - **限制**：如果无论选择哪种方向都无法使两路口类型不同，则输出“GG”。

2. **数据表示**：
    - **路口连接**：使用两个数组 `x` 和 `y` 分别记录每个路口在方向 X 和 Y 上连接到的下一个路口编号。
    - **路口类型**：使用数组 `t` 存储每个路口的类型（0 或 1）。

3. **模拟行走过程**：
    - **分别模拟两种方向**（X 和 Y）：
        - 对于每种方向，分别从路口 `A` 和 `B` 出发，按照选定的方向同步移动。
        - 在每一步，检查当前 `A` 和 `B` 所在路口的类型是否不同。
        - 如果不同，记录当前步数作为“路口独特度”。
        - 如果回到之前的路口组合（即进入循环），且类型始终相同，则该方向无法达成目标。

4. **循环检测**：
    - 使用集合 `visited` 记录已经访问过的 `(current_A, current_B)` 状态，以避免无限循环。
    - 如果在模拟过程中再次遇到相同的 `(current_A, current_B)` 状态，说明进入了循环，此方向无法达成类型不同的目标。

5. **结果比较**：
    - 对两种方向（X 和 Y）的模拟结果进行比较，选择能够达到目标的最小步数。
    - 如果两种方向都无法达成目标，则输出“GG”。

### **详细步骤**

1. **读取输入**：
    - 读取测试用例数量 `T`。
    - 对于每个测试用例，读取路口数量 `n`，起点路口 `A` 和 `B`。
    - 读取每个路口的连接信息和类型，填充数组 `x`、`y` 和 `t`。

2. **模拟函数** `simulate(direction)`：
    - **初始化**：步数 `step = 0`，当前路口 `current_A = A`，`current_B = B`，空集合 `visited`。
    - **循环**：
        - 检查 `current_A` 和 `current_B` 的类型是否不同，若不同，返回当前步数。
        - 检查当前状态 `(current_A, current_B)` 是否已访问，若已访问，返回“GG”。
        - 记录当前状态到 `visited`。
        - 根据方向 `X` 或 `Y` 更新 `current_A` 和 `current_B` 到下一个路口。
        - 步数 `step` 增加。
    - **返回**：“GG” 或找到的步数。

3. **处理每个测试用例**：
    - 分别调用 `simulate('X')` 和 `simulate('Y')`。
    - 比较两种方向的结果，选择最小的步数作为最终结果。
    - 如果两种方向都返回“GG”，则输出“GG”。

4. **输出结果**：
    - 对每个测试用例，输出计算得到的“路口独特度”或“GG”。

### **优化考虑**

- **高效循环检测**：使用集合 `visited` 快速检测是否进入循环，避免不必要的模拟步数。
- **分方向独立模拟**：分别模拟 X 和 Y 两种方向，独立判断其可行性，确保全面覆盖所有可能性。
- **早期退出**：一旦在某个方向找到满足条件的步数，立即记录，不再继续无效的步数计算。

### **总结**

该算法通过分别模拟两种可能的行走方向，利用循环检测机制有效地判断是否能够在有限步数内使两路口的类型不同，从而计算出“路口独特度”。这种方法确保了算法的正确性和效率，能够处理大规模的数据输入。
