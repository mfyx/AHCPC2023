```python
import math

# 定义函数，计算最小值
def calculate_min_value(a, b):
    # 根据给出的表达式 x = (sqrt(a) * b) / (sqrt(a) + 1)
    x = (math.sqrt(a) * b) / (math.sqrt(a) + 1)
    
    # 根据公式 f(x) = sqrt(x^2 + a) 和 g(x) = sqrt((b - x)^2 + 1)
    f_x = math.sqrt(x**2 + a)
    g_x = math.sqrt((b - x)**2 + 1)
    
    # 返回 f(x) + g(x)
    return f_x + g_x

# 读取输入
a, b = map(float, input().split())

# 计算并输出结果，保留小数点后 6 位
result = calculate_min_value(a, b)
print(f"{result:.6f}")
```
